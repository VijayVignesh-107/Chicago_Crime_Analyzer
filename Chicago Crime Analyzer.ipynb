```python
#Filling the null values of Latitude using the groupby method.
fill_latitude_1 = df.groupby(['District','Block'])['Latitude'].transform('median')
df['Latitude'] = df['Latitude'].fillna(fill_latitude_1)

fill_latitude_2 = df.groupby(['District'])['Latitude'].transform('median')
df['Latitude'] = df['Latitude'].fillna(fill_latitude_2)
```

```python
#Filling the null values of Longitude using the groupby method and median.
fill_Longitude_1 = df.groupby(['District','Block'])['Longitude'].transform('median')
df['Longitude'] = df['Longitude'].fillna(fill_Longitude_1)

fill_Longitude_2 = df.groupby(['District'])['Longitude'].transform('median')
df['Longitude'] = df['Longitude'].fillna(fill_Longitude_2)
```

```python
#Filling the null values of X Coordinate using the groupby method.
fill_XCoordinate_1 = df.groupby(['District','Block'])['X Coordinate'].transform('median')
df['X Coordinate'] = df['X Coordinate'].fillna(fill_XCoordinate_1)

fill_XCoordinate_2 = df.groupby(['District'])['X Coordinate'].transform('median')
df['X Coordinate'] = df['X Coordinate'].fillna(fill_XCoordinate_2)
```

```python
#Filling the null values of Y Coordinate using the groupby method.
fill_YCoordinate_1 = df.groupby(['District','Block'])['Y Coordinate'].transform('median')
df['Y Coordinate'] = df['Y Coordinate'].fillna(fill_YCoordinate_1)

fill_YCoordinate_2 = df.groupby(['District'])['Y Coordinate'].transform('median')
df['Y Coordinate'] = df['Y Coordinate'].fillna(fill_YCoordinate_2)
```

```python
##Filling the null values of Ward using the groupby method.
fill_Ward_1 = df.groupby(['District','Beat'])['Ward'].transform('median')
df['Ward'] = df['Ward'].fillna(fill_Ward_1)

fill_Ward_2 = df.groupby(['District'])['Ward'].transform('median')
df['Ward'] = df['Ward'].fillna(fill_Ward_2)
```

```python
#Filling the null values of Community Area using the groupby method.
fill_CommunityArea_1 = df.groupby(['District','Ward'])['Community Area'].transform('median')
df['Community Area'] = df['Community Area'].fillna(fill_CommunityArea_1)

fill_CommunityArea_2 = df.groupby(['District'])['Community Area'].transform('median')
df['Community Area'] = df['Community Area'].fillna(fill_CommunityArea_2)
```

```python
#Filling the null values of location by concatinating the latitude and longitude filled in previous cells.
df['Location'] = '(' + df['Latitude'].astype(str) + ',' + df['Longitude'].astype(str) + ')'
```

```python
# #Filling the null values of Location Description using the groupby method.
df['Location Description'] = df['Location Description'].astype(str)
fill_loc_description=df.groupby(['Primary Type', 'Description'])['Location Description'].apply(lambda x: x.mode()[0] if not x.mode().empty else pd.NA)
df['Location Description'] = df.apply()
```
